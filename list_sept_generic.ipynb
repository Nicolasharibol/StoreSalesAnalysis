```python
import pandas as pd
```

# September Sales Data 

```python
df = pd.read_excel('/Users/admin/Downloads/Monatsreport /list_sept_generic.xlsx')
df
```

```python
df.columns
```

```python
df = df[df['Item'] != 'opening_closing']
```

```python
df = df.drop(['Type','Date','Store','Pos','Number','Line','Article Code','Client','Retailer','Amount VAT','Mode Creation','Mode Generation','Tax Family','Document Number','Mode Creation2','name/surname','Total unit excluding VAT'], axis=1)
df
```

```python
df.info
```

```python
df['Occurrences'] = df.groupby('Item')['Item'].transform('size')
df
```

```python
df['Total Quantity'] = df.groupby('Item')['Quantity'].transform('sum')
df
```

# Top 20 Items - Quantity Sold

```python
top_20_items_quantity = (df.groupby('Item')['Quantity'].sum().reset_index().sort_values(by='Quantity', ascending=False).head(20).reset_index(drop=True))
top_20_items_quantity.index = top_20_items_quantity.index + 1
top_20_items_quantity
```

# Top 20 Items - Price unit 

```python
top_20_items_value = (df.assign(IndividualTotalValue=lambda x: x['Quantity'] * x['Price unit including VAT'])
    .groupby('Item', as_index=False)
    .agg({'Quantity': 'sum', 'IndividualTotalValue': 'sum'})
    .rename(columns={'Quantity': 'Quantity', 'IndividualTotalValue': 'Total Value'})
    .sort_values(by='Total Value', ascending=False)
    .reset_index(drop=True)
    .head(20))

top_20_items_value['Total Value'] = top_20_items_value['Total Value'].apply(lambda x: f'€{x:,.2f}')
top_20_items_value.index = top_20_items_value.index + 1
top_20_items_value
```

# Top 20 Items - Net Value

```python
top_20_items_value_netto = (df.assign(IndividualTotalValue=lambda x: x['Quantity'] * x['Price unit netto including VAT'])
    .groupby('Item', as_index=False)
    .agg({'Quantity': 'sum', 'IndividualTotalValue': 'sum'})
    .rename(columns={'Quantity': 'Quantity', 'IndividualTotalValue': 'Total Net Value'})
    .sort_values(by='Total Net Value', ascending=False)
    .reset_index(drop=True)
    .head(20))

top_20_items_value_netto['Total Net Value'] = top_20_items_value_netto['Total Net Value'].apply(lambda x: f'€{x:,.2f}')
top_20_items_value_netto.index = top_20_items_value_netto.index + 1
top_20_items_value_netto
```
